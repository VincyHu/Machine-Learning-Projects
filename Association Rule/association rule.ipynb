### Running Algorithm

```python
# Import packages
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# Set the column display to show all the columns 
pd.set_option('display.max_columns', None) 
```

```python
# import data
trans = pd.read_excel('transactions_by_dept.xlsx')
trans
```

```python
trans.shape # learn about the dataset size: there are 4539 records of transaction 
            # (Note that one transaction will have multiple columns if there are multiple items in one transaction)
```

Before calculating the antecedents and consequents, we need to consolidate the items into one transsaction per row with each product.

```python
basket = trans.groupby(['POS Txn','ID'])['Sales U'].sum().unstack().reset_index().fillna(0).set_index('POS Txn')
basket.head(5)
```

```python
# set all the value with >1 as a, set all other value as 0 in the basket dataframe
def set_quantity(x):
    if x>= 1:
        return 1
    if x<=0:
        return 0

basket_std = basket.applymap(set_quantity)
basket_std.head(5)
```

We will use **apriori** algorithm to build frequent itemsets.   
**Frequent itemset** is itemset that support is larger than the threshold set by marketing/sales team in the company.  
In **apriori** concept, frequent itemset are itemset that any combination of item in the itemset is also frequent itemset.The algorithm will iterate until excluding all non-frequent itemset in the itemsets.

```python
# build frequent itemsets
frequent_itemsets = apriori(basket_std,min_support=0.03,use_colnames=True)
```

```python
# display
frequent_itemsets
```

To further filter the result from so many frequent itemset, we will use **lift** as the final filtering metrics.  
**Lift** measures how strong the association rule is. When the lift value is larger than 1, then it indicates that the association have more probability than randomness of the two items.

```python
rules = association_rules(frequent_itemsets,metric = 'lift', min_threshold=1)
rules
```
